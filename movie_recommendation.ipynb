# As we can see above, we have a string containing a list of dictionaries. To convert this into a list we use the ast.literal_eval() function to convert string to list

```python
import ast
def extract_features(obj):
    li_of_features = []
    for i in ast.literal_eval(obj):
        li=i['name']
        li_of_features.append(li)
    return li_of_features
```

# Now,apply this function to the genres column

```python
movies['genres'].apply(extract_features)
```

```python
movies['genres'] = movies['genres'].apply(extract_features)
```

# iii) Now, we'll deal with the cast column

```python
movies['cast'][0]
```

```python
def extract_features_2(obj):
    li_of_features = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt!=3:
            li_of_features.append(i['name'])
            cnt+=1
        else:
            break
    return li_of_features
```

```python
movies['cast'].apply(extract_features_2)
```

```python
movies['cast'] = movies['cast'].apply(extract_features_2)
```

```python
movies.head()
```

# iv) Now, let's deal with the crew column, we wish to fetch only the director name 

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            director.append(i['name'])
    return director
    
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

 # Let's perform count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')                            #max_features = no_of_words to take
```

```python
#by default our function will return a scipy sparse matrix,we convert it to numpy array
cv.fit_transform(new_df['tags']).toarray()    
```

```python
cv.get_feature_names()
```

# We can observe that some words have their plurals as seperate features, to avoid this we need to perform stemming.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    l = []
    for word in text.split():
        l.append(ps.stem(word))
    return " ".join(l)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Now, we can continue with count vectorization

```python
vectors = cv.fit_transform(new_df['tags']).toarray()  
vectors[0]
```

```python
cv.get_feature_names()
```

# We need to find cosine similarity for vectors representing movies (array of arrays)

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend_movies(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(i[0])
```

# Final function for our recommendation system

```python
#final function
def recommend_movies(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for movie in movies_list:
        print(new_df.iloc[movie[0]].title)
```

```python
recommend_movies('Avatar')
```

```python
recommend_movies('The Dark Knight')
```
